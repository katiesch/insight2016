For graffiti and dumping reports, determine the distance to the nearest streetlight.

```python
sl_reader = sf.Reader('shapefiles/Street_Light/Street_Light.shp')
sl_record = pd.DataFrame(sl_reader.records())
```

```python
sl_shapes = sl_reader.shapes()
```

```python
# Need to convert each Street Light location to latitude and longitude
sl_record['lon'] = None
sl_record['lat'] = None
```

```python
p1 = Proj("+init=EPSG:2230", preserve_units=True)
print len(sl_record)
for i in range(0,len(sl_record)): 
    lon, lat = p1(sl_shapes[i].points[0][0], sl_shapes[i].points[0][1],inverse=True)
    sl_record.loc[i, 'lon'] = lon 
    sl_record.loc[i, 'lat'] = lat
    if (i % 2000==0):
        print i, i, i 
```

```python
#sl_record.to_csv('output_files/streetlight_latlon.csv', index=False)
```

For each event, I want to calculate the minimum distance from a street light. 
Only doing this for graffiti and dumping.

```python
requests_unique['sl_mindist'] = None
```

```python
for i in range(0,len(requests_unique)): 
    if ((requests_unique.loc[i,'Problem Category']=='Graffiti') 
        | (requests_unique.loc[i, 'Problem Category']=='Litter/Dumping')): 
        lat1=requests_unique.loc[i, 'Geolocation (Latitude)']
        lon1=requests_unique.loc[i, 'Geolocation (Longitude)']

        R = 6371 * 3280.84 # feet

        dLat = np.radians(np.array(np.array(sl_record['lat']) - lat1).astype(float))
        dLon = np.radians(np.array(np.array(sl_record['lon']) - lon1).astype(float))
        lat_source = np.radians(lat1)
        lat_comp = np.radians(np.array(sl_record['lat']).astype(float))


        a = np.sin(dLat/2.0) * np.sin(dLat/2.0) + np.sin(dLon/2.0) * np.sin(dLon/2.0) * np.cos(lat_source) * np.cos(lat_comp) 
        c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
        d = R * c

        requests_unique.loc[i, 'sl_mindist'] = float(d[np.argmin(d)])
    
```

```python
requests_unique['sl_mindist'] = requests_unique['sl_mindist'].astype(float)
```

Now I have information about dumping, graffiti, street lights, etc. all organised by Census block groups. However, they are skewed by population estimates, report numbers. 

I need to process the counts such that they are not biased towards areas with large populations or high reporting rates.
